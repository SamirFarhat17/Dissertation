```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
dates = range(0,365)

bitcoin_data_path = os.path.join(os.getcwd(),'Data', 'BTC_USD_2019-11-29_2020-11-28-CoinDesk.csv')
bitcoin = pd.read_csv(bitcoin_data_path)
btc_values = bitcoin['24h Open (USD)'].values

euro_data_path = os.path.join(os.getcwd(), 'Data','eurotodollar.csv')
euro = pd.read_csv(euro_data_path)
euro.mul(10000)
euro_values = euro['Open'].values *10000

gbp_data_path = os.path.join(os.getcwd(), 'Data','gbptodollar.csv')
gbp = pd.read_csv(gbp_data_path)
gbp.mul(10000)
gbp_values = gbp['Open'].values *10000

aud_data_path = os.path.join(os.getcwd(), 'Data','auddollar.csv')
aud = pd.read_csv(aud_data_path)
aud.mul(10000)
aud_values = aud['Open'].values *10000
```

```python
fig, bitcoin_stuff = plt.subplots(figsize =(15,7))
bitcoin_stuff.plot(dates,btc_values)
bitcoin_stuff.plot(dates,euro_values)
bitcoin_stuff.plot(dates,gbp_values)
bitcoin_stuff.plot(dates,aud_values)

bitcoin_stuff.set_ylim([0,20000])
bitcoin_stuff.set_xlim([0,365])

plt.legend(['BTC', 'EUR', 'GBP', 'AUD'], loc='lower right')
plt.xlabel('1 year period')
plt.ylabel('Price of asset in USD')
plt.title('Bitcoin vs 1 US Dollar and fiat currencies vs the 10,000 USD')

```
